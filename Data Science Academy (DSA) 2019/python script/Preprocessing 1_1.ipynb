```python
#Fungsi untuk Bulan Februari, Mei & Juni

def get_prep2(txt):
    
    row = [line.strip('\n').split(',') for line in txt]
    
    #memisahkan daging sapi, bawang putih dan lainnya
    sapi = [i for i in row if i[1]  == 'Daging Sapi']
    bp = [i for i in row if i[1]  == 'Bawang Putih Impor']
    bukan_sb = [i for i in row if (i[1] != ('Daging Sapi')  and i[1] !=('Bawang Putih Impor'))]
    
    #membuat data frame yang bukan sapi dan bawang putih
    df_bukan_sb = pd.DataFrame(bukan_sb[1:],columns=['TANGGAL','KOMODITI','SATUAN','HARGA'])
    
    #formatting data daging sapi
    ##membuat list daging sapi
    sapi_fix = [[i[0],i[1]+i[2]+i[4],i[5], i[6]] if i[3]==' Daging' else [i[0],i[1]+i[2]+i[3],i[4],i[5]] for i in sapi]
    ##membuat data frame daging sapi
    df_sapi = pd.DataFrame(sapi_fix,columns=['TANGGAL','KOMODITI','SATUAN','HARGA'])
    
    #formatting data bawang putih
    ##membuat list bawang putih
    bp_fix = [[i[0],i[1]+i[2],i[3],i[4]] for i in bp]
    ##membuat data frame bawang putih
    df_bp = pd.DataFrame(bp_fix,columns=['TANGGAL','KOMODITI','SATUAN','HARGA'])
    #Append Data Bulan Februari

    ##menggabungkan data daging sapi, bawang putih dan bukan daging sapi/bawang putih menjadi dataframe Januari    
    bulan = df_bukan_sb.append(df_bp).append(df_sapi)
    
    #ubah index
    bulan.index = range(len(row)-1)
    
    #menghapus ',','(' & ')' pada KOMODITI
    bulan['KOMODITI'].replace(regex=True,inplace=True,
                      to_replace=[',','\)','\('],value=['','',''])
    
    #mengubah harga string ke float
    bulan['HARGA'] = bulan['HARGA'].str.replace(r'\D+', '').astype('float')
    
    #mengubah tanggal string to datetime
    bulan['TANGGAL'] = pd.to_datetime(bulan['TANGGAL'], format = '%d/%m/%Y')

    return(bulan)
```

```python
# Fungsi untuk Bulan Maret, September & Oktober

def get_prep3(bulan):
    
    #menyamakan nama kolom
    bulan.columns = ['TANGGAL','KOMODITI','SATUAN','HARGA']
    
    #menghapus ',','(' & ')' pada KOMODITI
    bulan['KOMODITI'].replace(regex=True,inplace=True,
                      to_replace=[',','\)','\('],value=['','',''])
    
    #mengubah harga string ke float
    bulan['HARGA'] = bulan['HARGA'].astype('float')
        
    #mengubah tanggal string to datetime
    bulan['TANGGAL'] = pd.to_datetime(bulan['TANGGAL'], format = '%d/%m/%Y')
    
    return(bulan)
```

```python
# Fungsi untuk Bulan April, Juli, Agustus

def get_prep3k(bulan):
    
    #menyamakan nama kolom
    bulan.columns = ['TANGGAL','KOMODITI','SATUAN','HARGA']
    
    #menghapus ',','(' & ')' pada KOMODITI
    bulan['KOMODITI'].replace(regex=True,inplace=True,
                      to_replace=[',','\)','\('],value=['','',''])
    
    #mengubah harga string ke float
    bulan['HARGA'] = bulan['HARGA'].astype('float')
    
    #mengubah tanggal string to datetime
    bulan['TANGGAL'] = pd.to_datetime(bulan['TANGGAL'], format = '%Y-%m-%d')
    
    return(bulan)
```

```python
# Fungsi untuk Bulan November & Desember

def get_prep4(data,M,m):
    data.columns = ['TANGGAL','BULAN','TAHUN','KOMODITI','SATUAN','HARGA']
    df = data.dropna()
    
    #mengubah tanggal float ke string
    df.TANGGAL = df.TANGGAL.astype(int)
    df.TANGGAL = df.TANGGAL.astype(str)
    
    df['KOMODITI'].replace(regex=True,inplace=True,to_replace=[',','\)','\('],value=['','',''])
    
    #mengubah bulan huruf ke bulan angka
    df['BULAN'].replace(regex=True,inplace=True,to_replace=M,value=m)
    #mengubah bulan float ke string
    df.BULAN = df.BULAN.astype(int)
    df.BULAN = df.BULAN.astype(str)
    
    #mengubah tahun float ke string
    df.TAHUN = df.TAHUN.astype(int)
    df.TAHUN = df.TAHUN.astype(str)
    
    #membuat kolom baru dengan format tanggal/bulan/tahun
    df['TANGGAL'] = df['TANGGAL'].str.cat(df['BULAN'],sep='/')
    df['TANGGAL'] = df['TANGGAL'].str.cat(df['TAHUN'],sep='/')
    del df['BULAN']
    del df['TAHUN']
    
    #mengubah harga string to integer
    df['HARGA'] = df['HARGA'].astype('str')
    df['HARGA'] = df['HARGA'].str.replace(r'\D+', '').astype('float')

    #mengubah tanggal string to datetime
    df['TANGGAL'] = pd.to_datetime(df['TANGGAL'], format = '%d/%m/%Y')
    
    return(df)
```

```python
# Fungsi untuk Bulan November & Desember

def get_prep5(data,M,m):
    data.columns = ['TANGGAL','BULAN','TAHUN','KOMODITI','SATUAN','HARGA']
    df = data.dropna()
    
    #mengubah tanggal float ke string
    df.TANGGAL = df.TANGGAL.astype(int)
    df.TANGGAL = df.TANGGAL.astype(str)
    
    df['KOMODITI'].replace(regex=True,inplace=True,to_replace=[',','\)','\('],value=['','',''])
    
    #mengubah bulan huruf ke bulan angka
    df['BULAN'].replace(regex=True,inplace=True,to_replace=M,value=m)
    #mengubah bulan float ke string
    df.BULAN = df.BULAN.astype(int)
    df.BULAN = df.BULAN.astype(str)
    
    #mengubah tahun float ke string
    df.TAHUN = df.TAHUN.astype(int)
    df.TAHUN = df.TAHUN.astype(str)
    
    #membuat kolom baru dengan format tanggal/bulan/tahun
    df['TANGGAL'] = df['TANGGAL'].str.cat(df['BULAN'],sep='/')
    df['TANGGAL'] = df['TANGGAL'].str.cat(df['TAHUN'],sep='/')
    del df['BULAN']
    del df['TAHUN']
    
    #mengubah harga string to integer
    df['HARGA'] = df['HARGA'].astype('float')
    df['HARGA'] = round(df['HARGA'],2)
    #mengubah tanggal string to datetime
    df['TANGGAL'] = pd.to_datetime(df['TANGGAL'], format = '%d/%m/%Y')
    
    return(df)
```
